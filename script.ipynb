```python
import pandas as pd
import numpy as np

cereal = pd.read_csv("cereal.csv")

cereal.head()
```

```python
cereal.sample(5)
```

```python
cereal.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ax = plt.subplots()
ax = sns.distplot(cereal['calories'], kde=True, hist=True, bins=15)
ax.set_title('Cereal Calories')
ax.set_xlabel('calories')
plt.show()
```

```python
ax = plt.subplots()
ax = sns.distplot(cereal['vitamins'], kde=True, hist=True, bins=15)
ax.set_title('Cereal Vitamins')
ax.set_xlabel('vitamins')
plt.show()
```

```python
cereal.corr()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16, 5))
sns.heatmap(cereal.corr(), ax=ax[0])
ax[0].set_title('Basic Correlation Heatmap', pad=12)
sns.heatmap(cereal.corr(), vmin=-1, vmax=1, annot=True, ax=ax[1])
ax[1].set_title('Correlation Heatmap with Number', pad=12)
plt.show()
```

```python
ax2 = plt.subplots()
ax2 = sns.regplot(x='sugars', y='calories', data=cereal, fit_reg=True)
ax2.set_title('Sugars vs Calories of Cereals')
ax2.set_xlabel('sugars')
ax2.set_ylabel('calories')
plt.show()
```

```python
import seaborn.objects as so

filtered_df = cereal[(cereal['rating'] >= 50)]
fig, ax = plt.subplots(figsize=(10, 7))
p = (
    so.Plot(filtered_df, x='type', y='rating')
    .add(so.Bar())
    .label(
        x='Type',
        y='Rating',
        title='Type and Rating'
    )
)
p.on(ax).plot()
plt.show()
```
